```python
#2. masukkan kolom kedalam varibel
X = df['luas tanah'].values.reshape(-1,1)
y = df['jumlah panen padi '].values.reshape(-1,1)
#3. split data u/ training 70% dan u/ testing 30%
from sklearn.model_selection import train_test_split

x_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#4. analisis regresi
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
import pandas as pd

def check_missing(data, output_path=None):

    result = pd.concat([data.isnull().sum(), data.isnull().mean()], axis=1)
    result = result.rename(index=str, columns={0: 'total missing', 1: 'proportion'})

    if output_path is not None:
        result.to_csv(output_path + 'missing.csv')
        print('Result saved at', output_path, 'missing.csv')

    return result

check_missing(df)

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = {
    'luas tanah': [199431, 184029, 193666, 215975, 190608, 178766, 178307, 171395, 164826, 165729, 159463, 156803, 154941, 140613, 149888, 143034, 155802, 153897, 157441, 149369, 153243, 145990, 122214, 96588, 97690, 86202.68, 69536.06, 84772.93],
    'jumlah panen padi': [607529, 564722, 593502, 657650, 572269, 550277, 553641, 536779, 556564, 561007, 578346, 579404, 579635, 544597, 58663, 581704, 644947, 628828, 646641, 625164, 664535, 664720, 541486, 642095, 678127, 383045.74, 309932.68, 374376.27]
}
df = pd.DataFrame(data)

# Memisahkan fitur (x) dan target (y)
x = df.drop(columns='jumlah panen padi')
y = df['jumlah panen padi']

# Memisahkan data menjadi data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

lin_reg = LinearRegression()

# Melatih model
lin_reg.fit(x_train, y_train)

print("Koefisien:", lin_reg.coef_)
print("Intersepsi:", lin_reg.intercept_)

# Membuat DataFrame untuk koefisien
coef_dict = {
    'Fitur': x.columns,
    'Koefisien': lin_reg.coef_
}

coef = pd.DataFrame(coef_dict, columns=['Fitur', 'Koefisien'])
print(coef)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Prediksi nilai menggunakan data uji
y_pred = lin_reg.predict(x_test)

# Menghitung MSE dan R^2
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R^2):", r2)
```

nilai R-squared adalah 0.39147712009689706. Ini menunjukkan bahwa model dapat menjelaskan sekitar 39.15% dari variabilitas data, sedangkan sisanya (sekitar 60.85%) tidak dapat dijelaskan oleh model. Semakin mendekati nilai 1, semakin baik model dalam menjelaskan data.

Secara keseluruhan, hasil yang diperoleh menunjukkan bahwa model tidak sepenuhnya cocok dengan data, tetapi masih memberikan sejumlah informasi yang berguna dalam menjelaskan variabilitas data
